```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import countDistinct
```

```python
spark = SparkSession.builder \
    .appName("NetworkTrafficAnomaliesDetection") \
        .getOrCreate()
```

```python
spark
```

```python
# normalize column names by removing leading/trailing spaces
df_friday_ddos = spark.read.csv('../data/raw/Friday-WorkingHours-Morning.pcap_ISCX.csv', header=True, inferSchema=True)
df_friday_ddos = df_friday_ddos.toDF(*[c.strip() for c in df_friday_ddos.columns])
df_friday_ddos.printSchema()
df_friday_ddos.select('Label').distinct().show()
df_friday_ddos.groupBy("Label").count().orderBy("count").show()
df_friday_ddos.where(F.col('Label') != 'BENIGN').count()
```
